## Demo 2: look up words in jmdict_english.zip 
https://foosoft.net/projects/yomichan/

```python
# https://dev.to/mathewthe2/making-a-japanese-dictionary-lookup-tool-with-sudachi-in-python-4k9i

from sudachipy import tokenizer
from sudachipy import dictionary

tokenizer_obj = dictionary.Dictionary(dict_type='small').create()
mode = tokenizer.Tokenizer.SplitMode.A
```

```python
from pathlib import Path
import zipfile
import json

SCRIPT_DIR = '/Users/kotaminegishi/dictionaries'
dictionary_map = {}

def load_dictionary(dictionary):
    output_map = {}
    archive = zipfile.ZipFile(dictionary, 'r')

    result = list()
    for file in archive.namelist():
        if file.startswith('term'):
            with archive.open(file) as f:
                data = f.read()  
                d = json.loads(data.decode("utf-8"))
                result.extend(d)

    for entry in result:
        if (entry[0] in output_map):
            output_map[entry[0]].append(entry) 
        else:
            output_map[entry[0]] = [entry] # Using headword as key for finding the dictionary entry
    return output_map

dictionary_map = load_dictionary(str(Path(SCRIPT_DIR, 'jmdict_english.zip')))
```

```python
def look_up(word):
    word = word.strip()
    if word not in dictionary_map:
        m = tokenizer_obj.tokenize(word, mode)[0]
        word = m.dictionary_form()
        if word not in dictionary_map:
            return None
    result = [{
        'headword': entry[0],
        'reading': entry[1],
        'tags': entry[2],
        'glossary_list': entry[5],
        'sequence': entry[6]
    } for entry in dictionary_map[word]]
    return result
```

```python
print(look_up('牛丼'))
print(look_up('食べて'))
print(look_up('自己紹介'))
```

## Exploring custom functions

```python
import re

def token_lookup(line, verbose=False):   
    tokens = tokenizer_obj.tokenize(line, mode)
    line_tokens = '_'.join([str(i) for i in tokens])
    if verbose: print('tokens:', line_tokens)

    words = [token.dictionary_form() for token in tokens]
    words = [word for word in words if len(word)>1] # ignore single-character tokens
    word_dict = {}
    for word in words:
        if verbose: print(f'word to look up: {word}')
        result = jam.lookup(word)

        if result.entries != None:
            word_meanings = ''
            for entry in result.entries:
                word_meanings += re.sub('<bound method JMDEntry.text of |>', '',str(entry.text)) + '\n'
                if verbose: print(word_meanings) 
            word_dict.update({word: word_meanings})             

    return line_tokens, word_dict
```

```python
remove_vocab = ['メイ', 'ある', 'こう','10','11','12','13','14','15','16','17','18']
vocab = pd.Series(list_word_dict).apply(lambda x: [key for key in x if (x.get(key)!='') & (not(key in remove_vocab))])
```

```python
# spot unnecessary words and add them to remove_vocab
vocab_list = []
for row in vocab:
    vocab_list += row 

vocab_list = pd.Series(vocab_list).unique()
vocab_list
```

```python
df_lines['vocabulary'] = vocab
df_lines.head()
```

```python
import re

ignore_tokens1 = ['。','、','？','！','…']

def remove_quote(line):
    quo_beg = line.find('「') + 2 if line.find('「') >0 else 0 
    quo_end = line.find('」') - 1 if line.find('」') >0 else len(line) 
    line_extract = line[quo_beg:quo_end]
    for char in ignore_tokens1:
        line_extract = re.sub('_'+char,char,line_extract)
        line_extract = re.sub(char+'_',char,line_extract)
    return line_extract

remove_quote(df_lines['tokens'][1]), remove_quote(df_lines['tokens'][15])
```
